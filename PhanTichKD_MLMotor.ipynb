###1/ Vẽ chart từ số lượng xe các tỉnh ngoài HCM trong từng tháng (Trục tung SL , trục hoành Tháng ) ?

```python
import matplotlib.pyplot as plt
tinhLe = df1[df1['Tỉnh'] != 'Hồ Chí Minh']
tinhLe.groupby([df1['Ngày bán'].dt.day])['STT'].count().plot(kind='bar', figsize=(12,8))

plt.xticks(rotation=0)
plt.xlabel('Ngày')
plt.ylabel('SL', rotation=0)
plt.title('Số lượng xe tỉnh ngoài trong từng tháng')
```

```python

demSL_Ngay = tinhLe.groupby([df1['Ngày bán'].dt.date])['STT'].count().rename({'Ngày bán':'STT'}, axis='columns')
print(demSL_Ngay)
```

```python
## 2/ Vẽ chart Ngày (Cụ thể Thứ ngày) nào trong năm có SL xe bán nhiều hơn 5 xe ?
perDay = df1.groupby(df1['Ngày bán'].dt.day).STT.count().to_frame().reset_index()
perDay = perDay.rename({'STT':'SL'}, axis='columns')
perDay.loc[perDay['SL'] > 5]
```

```python
import matplotlib.pyplot as plt

perDay.loc[perDay['SL'] > 5].plot(kind='bar', figsize=(12,8))

plt.xticks(rotation=0)
plt.xlabel('Ngày')
plt.ylabel('SL', rotation=0)
plt.title('Ngày trong năm có SL bán nhiều hơn 5 xe')
```

```python
#Doanh số từng tháng trong năm
doanhSo = df1.groupby([df1['Ngày bán'].dt.month])['Tiền thanh toán'].sum().to_frame()
doanhSo = doanhSo.reset_index().rename({'Ngày bán':'tháng', 'Tiền thanh toán':'doanh thu'}, axis='columns')
doanhSo['doanh thu'] = doanhSo['doanh thu']/1000000000
plt.figure(figsize=(12, 8))
plt.bar(doanhSo['tháng'], doanhSo['doanh thu'])
plt.xticks(range(1, 13))
plt.ylabel('doanh số\n(tỷ vnđ)', rotation=0)
plt.title('Doanh số từng tháng')
```

```python
## 4/ Xếp hạng 10 xe và màu sắc của xe đó bán chạy nhất trong năm ?
banChay = df1.groupby([df1['Loại xe'], df1['Màu xe']]).STT.count().sort_values(ascending=False).to_frame().reset_index()
banChay['Loại, màu xe'] = banChay['Loại xe'] + '\n' + banChay['Màu xe']
top10 = banChay.iloc[:10, :]
```
