```python
import pandas as pd
```

```python
yamaha_danhgiakh = pd.read_excel("yamaha_danhgiakh.xls", usecols=[1,2,3,4,5,6])
yamaha_danhgiakh
```

```python
# chuyển cột DateCreated từ Series -> datetime
yamaha_danhgiakh['DateCreated'] = pd.to_datetime(yamaha_danhgiakh['DateCreated'])

# tạo từng cột: ngày, tháng, năm, giờ, phút, giây
yamaha_danhgiakh['day'] = yamaha_danhgiakh['DateCreated'].dt.day
yamaha_danhgiakh['month'] = yamaha_danhgiakh['DateCreated'].dt.month
yamaha_danhgiakh['year'] = yamaha_danhgiakh['DateCreated'].dt.year
yamaha_danhgiakh['hour'] = yamaha_danhgiakh['DateCreated'].dt.hour
yamaha_danhgiakh['minute'] = yamaha_danhgiakh['DateCreated'].dt.minute
yamaha_danhgiakh['second'] = yamaha_danhgiakh['DateCreated'].dt.second
```
